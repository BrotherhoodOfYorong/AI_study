```python
import pandas as pd

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train.interpolate(inplace=True)
test.fillna(0, inplace=True)
```

```python
Y_train = train['count']
X_train = train.drop(['count', 'hour_bef_precipitation', 'hour_bef_pm2.5', 'hour_bef_windspeed'], axis=1)

test = test.drop(['hour_bef_precipitation', 'hour_bef_pm2.5', 'hour_bef_windspeed'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor(criterion = 'mse', random_state=2020)

params = {
    'n_estimators': [200, 300, 500],
    'max_features': [5, 6, 8],
    'min_samples_leaf': [1, 3, 5]
}

greedy_CV = GridSearchCV(model, param_grid=params, cv = 3, n_jobs = -1)
greedy_CV.fit(X_train, Y_train)
```

```python
result = greedy_CV.predict(test)
```

```python
result
```

```python
submission = pd.read_csv('data/submission.csv')
```

```python
submission['count'] = result
```

```python
submission.to_csv('sub_4.csv', index=False)
```
